```python
import importlib.util
import subprocess
import sys

if importlib.util.find_spec("ipywidgets") is None:
    subprocess.run(["uv", "add", "ipywidgets"], check=True)

import matplotlib.pyplot as plt
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

movies = netflix_titles[netflix_titles["type"] == "Movie"].copy()
movies["release_year"] = pd.to_numeric(movies["release_year"], errors="coerce")
movies = movies.dropna(subset=["release_year"])
movies["release_year"] = movies["release_year"].astype(int)

year_counts = movies.groupby("release_year").size().sort_index()
min_year = int(year_counts.index.min())
max_year = int(year_counts.index.max())

year_slider = widgets.IntRangeSlider(
    value=[max(min_year, max_year - 20), max_year],
    min=min_year,
    max=max_year,
    step=1,
    description="Years",
    continuous_update=False,
    layout=widgets.Layout(width="700px"),
)

top_n_slider = widgets.IntSlider(
    value=10,
    min=5,
    max=30,
    step=1,
    description="Top titles",
    continuous_update=False,
    layout=widgets.Layout(width="350px"),
)

sort_dropdown = widgets.Dropdown(
    options=[
        ("Newest first", "desc"),
        ("Oldest first", "asc"),
    ],
    value="desc",
    description="Order",
    layout=widgets.Layout(width="250px"),
)

output = widgets.Output()


def update_dashboard(*_):
    with output:
        clear_output(wait=True)

        start_year, end_year = year_slider.value
        filtered = movies[(movies["release_year"] >= start_year) & (movies["release_year"] <= end_year)].copy()

        if sort_dropdown.value == "desc":
            filtered = filtered.sort_values(["release_year", "title"], ascending=[False, True])
        else:
            filtered = filtered.sort_values(["release_year", "title"], ascending=[True, True])

        filtered_counts = filtered.groupby("release_year").size().reindex(
            range(start_year, end_year + 1), fill_value=0
        )

        fig, ax = plt.subplots(figsize=(12, 5))
        ax.bar(filtered_counts.index, filtered_counts.values, color="#E50914")
        ax.set_title("Netflix Movies by Release Year")
        ax.set_xlabel("Release year")
        ax.set_ylabel("Number of movies")
        ax.grid(axis="y", linestyle="--", alpha=0.3)
        plt.tight_layout()
        plt.show()

        summary = pd.DataFrame(
            {
                "metric": [
                    "Selected year range",
                    "Movies in range",
                    "Earliest movie in range",
                    "Latest movie in range",
                ],
                "value": [
                    f"{start_year}–{end_year}",
                    len(filtered),
                    filtered["release_year"].min() if not filtered.empty else "N/A",
                    filtered["release_year"].max() if not filtered.empty else "N/A",
                ],
            }
        )
        display(summary)

        display(filtered[["title", "release_year", "country", "rating"]].head(top_n_slider.value))


year_slider.observe(update_dashboard, names="value")
top_n_slider.observe(update_dashboard, names="value")
sort_dropdown.observe(update_dashboard, names="value")

controls = widgets.VBox([
    widgets.HTML("<h3>Explore year of creation of Netflix movies</h3>"),
    year_slider,
    widgets.HBox([top_n_slider, sort_dropdown]),
])

display(controls, output)
update_dashboard()
```
